3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cars_data=pd.read_csv("cars_data.csv")
sns.countplot(x="Buyer Gender", data=cars_data)
print('Number of Male buyers   : ',cars_data['Buyer Gender'].value_counts()['Male'])
print('Number of Female Buyers : ',cars_data['Buyer Gender'].value_counts()['Female'])
```
